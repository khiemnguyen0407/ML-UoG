```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1, x2 = np.linspace(-10, 10, 1001), np.linspace(-20, 20, 1001)
X = np.zeros(shape=(2, x1.shape[0], x2.shape[0]))
X[0], X[1] = np.meshgrid(x1, x2, indexing='ij')

a = np.array([50, 1/2])
f = a[0] * X[0]**2 + a[1] * X[1]**2

plt.figure(figsize=(5, 5))
plt.contour(X[0], X[1], f, levels=np.arange(1, 100, 10))
plt.grid(True)
```

```python
max_iter = 3
x = np.array([1, 1], dtype=np.float64)
alpha_list = [0.5]
for j in range(len(alpha_list)):
    alpha = alpha_list[j]
    x = np.array([0.5, 0.5])
    x_list = np.zeros(shape=(max_iter+1, 2))
    x_list[0,:] = x
    for i in range(max_iter):
        x[0] -= alpha * 2 * a[0] * x[0]
        x[1] -= alpha * 2 * a[1] * x[1]
        x_list[i+1,:] = x

    plt.figure(figsize=(16, 6), dpi=80)
    plt.subplot(121)
    plt.plot(x_list[:,0], 'k-o')
    plt.grid(True)
    plt.subplot(122)
    plt.plot(x_list[:,1], 'k-o')
    plt.grid(True)
    plt.title('$x_2$', fontsize=20)
    plt.show()

print(x_list)
```
